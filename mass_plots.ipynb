```python
import atmosphere as atmo
import aerodynamics as aero
import path
import rocket_parser as rp
import matplotlib.pyplot as plt
```

```python
parser = rp.rocket_parser(path.rocket_lib + "SHStarship.json")
```

```python
full_mass = parser.get_full_mass()
delta_mass = parser.get_delta_mass()
stage_mass = parser.get_stage_mass()
propellant_mass = parser.get_propellant_mass()
structural_mass = parser.get_structural_mass()
block_number = parser.get_block_number()
time_points = parser.get_work_time()
```

```python
plt.plot(parser.vector_time(), parser.vector_mass())
plt.title('Расчет параметра по времени полета', fontsize=16)
plt.xlabel('Время полета, с', fontsize=14)
plt.ylabel('Масса, кг', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.plot(parser.vector_time(), parser.vector_static())
plt.title('Расчет параметра по времени полета', fontsize=16)
plt.xlabel('Время полета, с', fontsize=14)
plt.ylabel('Статический момент Sx, кг*м2', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.plot(parser.vector_time(), parser.vector_inertia())
plt.title('Расчет параметра по времени полета', fontsize=16)
plt.xlabel('Время полета, с', fontsize=14)
plt.ylabel('Момент инерции Iz, кг*м3', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.plot(parser.vector_time()[:-1], parser.vector_center()[:-1])
plt.title('Расчет параметра по времени полета', fontsize=16)
plt.xlabel('Время полета, с', fontsize=14)
plt.ylabel('Центр масс, м', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
